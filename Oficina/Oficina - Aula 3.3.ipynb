```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = pd.read_csv('dadosoficina07.txt')
df.head()
```

# 1 - Vendas ao longo dos meses

> No gráfico abaixo é exibido as vendas mensais em um período de três anos. Podemos perceber:
> * um aumento de aproximadamente 100% das vendas entre o primeiro e o último mês.
> * no início do segundo ano as vendas estão abaixo do valor inicial do ano anterior.
> * a partir do segundo ano as vendas seguem aumentando, demonstrando um crescimento considerável.


```python
plt.figure(figsize=(20,5))
plt.title('Vendas mensais')
sns.lineplot(x='Month', y='Sales', data=df)
plt.show()
```
