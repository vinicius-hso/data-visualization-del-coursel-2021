## Análise de Filmes lançados entre 2007 e 2011

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes_uri = 'filmes.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

---
### 1 - Orçamento de cada filme:
>
>- __Maior orçamento:__ "Pirates of the Caribbean: At World's End"
>
>- __Menor orçamento:__ "The Twilight Saga: New Moon"

```python
plt.figure(figsize=(20, 15))
plt.title('Orçamento de cada filme')
sns.barplot(y='Film',
           x='Budget_M',
           ci=None,
           data=filmes)
sns.set(style='darkgrid')
plt.show()
```

---
### 2 -  Bilheteria de cada filme:

>- __Maior bilheteria__: "Avatar"
>
>- __Menor bilheteria__: "Ratatouille"

```python
plt.figure(figsize=(20, 15))
plt.title('Bilheteria de cada filme')
sns.barplot(y='Film',
           x='Worldwide_Gross_M',
           ci=None,
           data=filmes)
sns.set(style='darkgrid')
plt.show()
```

```python
# As 5 maiores bilheterias

filmes.sort_values(by='Worldwide_Gross_M', ascending=False).head()
```

```python
# As 5 menores bilheterias

filmes.sort_values(by='Worldwide_Gross_M').head()
```

----
### 3 - Quantidade de filmes por gênero em cada ano:

>- __Anos com menos filmes de ação__: 2008 e 2010
>
>- __Anos com menos filmes de animação__: 2008 e 2011

```python
plt.figure(figsize=(12, 5))
plt.title('Quantidade de filmes por gêneros em cada ano')
sns.countplot(x='Year', palette='muted', edgecolor='0.3', hue='Genre', data=filmes)
plt.show()
```

---
### 4 - Média de orçamento por gênero em cada ano:

>- __Ano de maior média para filmes de Ação__: 2007
>
>- __Ano de maior média para filmes de Animação__: 2008

```python
plt.figure(figsize=(13, 5))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x='Year', 
            y='Budget_M',
            palette='muted',
            edgecolor='0.3',
            hue='Genre', 
            ci='sd', 
            data=filmes)
plt.show()
```
