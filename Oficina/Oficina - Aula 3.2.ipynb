```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('mpg')
df.head()
```

## 1 - Relação entre 'Aceleração' e 'Consumo de combustível' de acordo com o 'País de origem':

> * __Origem__ = __USA__ - Podemos perceber uma __Baixa Correlação Positiva__
> * __Origem__ = __JAPAN__ - Podemos perceber uma __Correlação Inexistente__
> * __Origem__ = __EUROPE__ - Podemos perceber uma __Correlação Inexistente__

```python
sns.relplot(x='acceleration', y='mpg', col='origin', data=df)
plt.show()
```

### USA - Gráfico de Dispersão com Reta de Regressão

```python
usa = df[df['origin'] == 'usa']
plt.figure(figsize=(8,5))
plt.title('USA: Relação entre Aceleração e Consumo de Combustível')
sns.regplot(x='acceleration', y='mpg', data=usa)
plt.show()
```

### JAPAN - Gráfico de Dispersão com Reta de Regressão

```python
japan = df[df['origin'] == 'japan']
plt.figure(figsize=(8,5))
plt.title('JAPAN: Relação entre Aceleração e Consumo de Combustível')
sns.regplot(x='acceleration', y='mpg', data=japan)
plt.show()
```

### EUROPE - Gráfico de Dispersão com Reta de Regressão

```python
europe = df[df['origin'] == 'europe']
plt.figure(figsize=(8,5))
plt.title('EUROPE: Relação entre Aceleração e Consumo de Combustível')
sns.regplot(x='acceleration', y='mpg', data=europe)
plt.show()
```

## 2 - Relação entre 'Aceleração' e 'Consumo de Combustível' de acordo com 'Ano do Modelo':

> * Podemos perceber no Gráfico de Dispersão abaixo que, os modelos mais antigos (as cores mais claras) possuem uma taxa de consumo de combustível (mpg) mais baixa quando comparado com os modelos mais novos (cores mais escuras)

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre Aceleração e Consumo de Combustível de acordo com Ano do Modelo')
sns.scatterplot(x='acceleration', y='mpg', hue='model_year', data=df)
plt.show()
```
