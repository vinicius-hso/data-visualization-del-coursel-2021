```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='darkgrid')
```

# Titanic Dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## 1 - Distribuição do valor da passagem de acordo com a cidade de embarque:

```python
plt.figure(figsize=(8,5))
plt.title('Distribuição do valor da passagem de acordo com a cidade de embarque')
sns.violinplot(x='embark_town', y='fare', data=titanic)
plt.show()
```

## 2 - Distribuições do valor da passagem de acordo com:

> * __Cidade de Embarque__
> * __Sexo__
> * __Sobreviveu ou não__

> Pode-se perceber no Gráfico de Violino abaixo que, a maior parte dos __sobreviventes__:
> * embarcaram em __'Cherbourg'__ 
> * são do sexo __'Feminino'__
> * pagaram em média __$50__

> Pode-se notar também que, a maior parte dos __não sobreviventes__ :
> * pagaram um valor muito baixo pela passagem
> * embarcaram em 'Southampton'

```python
sns.catplot(x='embark_town',
           y='fare',
           hue='sex',
           col='survived',
           kind='violin',
           split=True,
           data=titanic)
plt.show()
```
