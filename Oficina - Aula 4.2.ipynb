```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('mpg')
df.head()
```

## 1 - Distribuição Conjunta de todas as variáveis:
> * Variáveis com maior correlação:
>>* horsepower X weight - Alta Correlação Positiva
>>* mpg X weight - Alta Correlação Negativa

> * Pode-se perceber que:
>>* Os carros produzidos nos USA são mais pesados e possuem maior força em "cavalos" (em sua maioria) quando comparados aos carros produzidos na Europa e Japão
>>* A curva de probabilidade das variáveis 'mpg', 'horsepower', 'weight' e 'aceleration' dos dados referentes ao USA se destacam mais que os demais, facilitando a identificação visual.

```python
# Crie uma visualização que mostre a distribuição conjunta de todas as variáveis 
# quantitativas, diferenciando pelo país de origem (‘origin’). Quais variáveis possuem 
# uma maior correlação?

'''
horsepower X weight
mpg X weight
'''

sns.pairplot(df, hue='origin')
plt.show()
```

## 2 - Distribuição Conjunta, Reta de Regressão e Distribuições Marginais

> * Gráfico __A__
>> * Correlação entre as variáveis: __'horsepower'__ e __'weight'__

```python
sns.jointplot(x='horsepower', y='weight', kind='reg', data=df)
plt.show()
```

> * Gráfico __B__
>> * Correlação entre as variáveis: __'mpg'__e __'weight'__

```python
sns.jointplot(x='mpg', y='weight', kind='reg', data=df)
plt.show()
```
